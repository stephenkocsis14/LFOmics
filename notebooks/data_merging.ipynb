# Combining Data Frames for Processing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load data
data_bRNAseq = pd.read_csv('bRNAseq_countMatrix.csv', index_col=0)
data_scRNAseq = pd.read_csv('scRNAseq_countMatrix.csv', index_col=0, usecols=range(50))
```

```python
data_bRNAseq
```

```python
# Rename the index column to 'GeneSymbol'
data_scRNAseq.rename_axis('GeneSymbol', inplace=True)
data_scRNAseq
```

```python
# Check column names in data_bRNAseq
print("Columns in data_bRNAseq:", data_bRNAseq.columns)

# Check column names in data_scRNAseq
print("Columns in data_scRNAseq:", data_scRNAseq.columns)
```

```python
# Reset index to make 'GeneSymbol' a column
data_bRNAseq_reset = data_bRNAseq.reset_index()
data_scRNAseq_reset = data_scRNAseq.reset_index()

# Check column names after resetting index
print("Columns in data_bRNAseq_reset:", data_bRNAseq_reset.columns)
print("Columns in data_scRNAseq_reset:", data_scRNAseq_reset.columns)

# Merge dataframes on column 'GeneSymbol'
merged_df = pd.merge(data_bRNAseq_reset, data_scRNAseq_reset, on='GeneSymbol')

# Reorder columns to have 'GeneSymbol' at the 0 index
merged_df = merged_df[['GeneSymbol'] + [col for col in merged_df.columns if col != 'GeneSymbol']]

merged_df
```

```python
# Export the merged dataframe to a CSV file
merged_df.to_csv('merged_dataframe.csv', index=False)
```
